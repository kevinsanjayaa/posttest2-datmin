#### Data Preparation


##### a. Select Data

```python
# Memilih kolom-kolom yang akan digunakan
selected_columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
df3_selected = df3[selected_columns]
```

##### b. Clean Data

```python
# Mengganti nilai-nilai 0 pada kolom Glucose, BloodPressure, SkinThickness, Insulin, dan BMI dengan nilai NaN
df3_cleaned = df3_selected.replace({'Glucose': 0, 'BloodPressure': 0, 'SkinThickness': 0, 'Insulin': 0, 'BMI': 0}, np.nan)

# Menghapus baris yang mengandung nilai NaN
df3_cleaned = df3_cleaned.dropna()
```

##### c. Construct Data

```python
# Menambahkan kolom baru 'Age_Group' berdasarkan umur pasien
df3_constructed = df3_cleaned.copy()
df3_constructed['Age_Group'] = pd.cut(df3_constructed['Age'], bins=[20, 30, 40, 50, 60, 70, 80], labels=['20-29', '30-39', '40-49', '50-59', '60-69', '70-79'])

# Menambahkan kolom baru 'BMI_Category' berdasarkan kategori BMI
df3_constructed['BMI_Category'] = pd.cut(df3_constructed['BMI'], bins=[0, 18.5, 25, 30, 100], labels=['Underweight', 'Normal', 'Overweight', 'Obese'])
```

##### d. Intergrated Data

```python
# Menggabungkan dataset diabetes dengan dataset lain yang berkaitan
df3_integrated = pd.concat([df3_constructed], axis=1)
```

##### e. Format Data

```python
# Mengubah tipe data kolom 'Glucose' menjadi float
df3_formatted = df3_integrated.astype({'Glucose': 'float'})
```
